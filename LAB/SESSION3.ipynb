```python
from matplotlib import pyplot as plt

x = [3,1,7,3,1,8,10,3,9,11,3,7,8,1,4,2]
y = [2,5,2,3,6,8,12,7,5,10,6,7,9,3,5,5]

plt.plot(x, y, '.')
plt.xlabel("Titik X")
plt.ylabel("Titik Y")
plt.title("Diagram XY")
plt.show()
```

```python
import numpy as np
x = np.array([3,1,7,3,1,8,10,3,9,11,3,7,8,1,4,2])
y = np.array([2,5,2,3,6,8,12,7,5,10,6,7,9,3,5,5])

m = 0.5
b = 3.3

plt.plot(x, y, '.')

y_new= m*x + b

# plot regression line
plt.plot(x, y_new, '-')
plt.xlabel("Titik X")
plt.ylabel("Titik Y")
plt.title("Diagram XY")
plt.show()
```

```python
A = np.vstack([x, np.ones(len(x))]).T

Y = np.vstack(y)

hasilInvers = np.linalg.inv(np.dot(A.T, A))
beta = np.dot(hasilInvers, np.dot(A.T, Y))
print(beta)

m = beta[0]
b = beta[1]

plt.plot(x, y, '.')

y_new= m*x + b

# plot regression line
plt.plot(x, y_new, '-')
plt.xlabel("Titik X")
plt.ylabel("Titik Y")
plt.title("Diagram XY")
plt.show()
```

```python
from scipy.interpolate import interp1d, CubicSpline

x = [1, 6, 11, 16]
y = [2, 5, 3, 7]

plt.plot(x, y, '.-')
plt.show()

x_cari = 4
function = interp1d(x, y, kind='cubic')
y_cari = function(x_cari)

plt.plot(x, y, 'o-')
plt.plot(x_cari, y_cari, 'ro')
plt.show()


# x_new = np.linspace(1, 16, num=41, endpoint=True)
# f = interp1d(x, y, kind='cubic')
# plt.plot(x_new, f(x_new), '-')
# plt.show()
```
