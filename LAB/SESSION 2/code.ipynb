```python
# Gauss Seidel 
# Diagonally Dominant
import numpy as np

a = [
    [6, 3, 2],
    [2, 7, 3],
    [7, 5, 15]
]

print(np.diag(a))
print(np.sum(np.abs(a), axis=1))

def checkDiagonallyDominant(a):
    diag = np.diag(a)
    sumWithDiag = np.sum(np.abs(a), axis=1)
    sumWithoutDiag = sumWithDiag - diag
    if np.all(diag > sumWithoutDiag):
        print("Diagonally dominant")
    else:
        print("Not diagonally dominant")

checkDiagonallyDominant(a)
```

```python
X = [
    [
    [-3,9,-6],
    [7,-2,-3],
    [4,1,-4]
    ],
  [
    [1, 3, -3, 9],
    [2, -9,3, -3],
    [-1, 3, 15, -1],
    [21, 4, -1, 10]
  ],
  [
    [4,2,-1],
    [1,3,-1],
    [5,2,10]
  ],
  [
    [-7,2,-4],
    [-1,-19,-13],
    [-3,3,-19]
  ],
  [
    [-12, 1, -4, 6],
    [-3, 8,-3, 2],
    [3, 3, 10, -3],
    [3, -2, 3, 12]
  ],
  [
    [2,1,-1],
    [-2,5,-3],
    [-1,-6,-9]
  ]
]
Y=[
   [0, -6, 5],
   [12, 9, -6, 11],
   [19, 12, 5],
   [8, 3, 5],
   [3, 6, 9, 21],
   [3, 9, 12]
]

def gaussSeidel(X, Y, epsilon, limit):
    n = len(X)
    x = np.zeros(n)
    for i in range(limit):
        for j in range(n):
            d = Y[j]
            for k in range(n):
                if(j != k):
                    d-=X[j][k] * x[k]
            x[j] = d/X[j][j]
        # print("Iteration {0}: {1}".format(i, x))
        if np.allclose(np.dot(X, x), Y, atol=epsilon):
            print("Converged")
            break
    
    return x

for i in range(len(X)):
    print("X: ", X[i])
    print("Y: ", Y[i])
    print("Solution: ", gaussSeidel(X[i], Y[i], 0.1, 15))
    print("")


```

```python
import numpy as np

# Define coefficient matrix and constant matrix
coefficientMatrix = [    [    [-3,9,-6],
    [7,-2,-3],
    [4,1,-4]
    ],
  [    [1, 3, -3, 9],
    [2, -9,3, -3],
    [-1, 3, 15, -1],
    [21, 4, -1, 10]
  ],
  [    [4,2,-1],
    [1,3,-1],
    [5,2,10]
  ],
  [    [-7,2,-4],
    [-1,-19,-13],
    [-3,3,-19]
  ],
  [    [-12, 1, -4, 6],
    [-3, 8,-3, 2],
    [3, 3, 10, -3],
    [3, -2, 3, 12]
  ],
  [    [2,1,-1],
    [-2,5,-3],
    [-1,-6,-9]
  ]
]
constantMatrix=[   [0, -6, 5],
   [12, 9, -6, 11],
   [19, 12, 5],
   [8, 3, 5],
   [3, 6, 9, 21],
   [3, 9, 12]
]

# Define the Gauss-Seidel method
def gaussSeidel(coefficientMatrix, constantMatrix, epsilon, limit):
    n = len(coefficientMatrix)
    x = np.zeros(n)
    for i in range(limit):
        for j in range(n):
            d = constantMatrix[j]
            for k in range(n):
                if(j != k):
                    d-=coefficientMatrix[j][k] * x[k]
            x[j] = d/coefficientMatrix[j][j]
        if np.allclose(np.dot(coefficientMatrix, x), constantMatrix, atol=epsilon):
            print("Converged")
            break
    
    return x

# Iterate through each matrix and print results
for i in range(len(coefficientMatrix)):
    print("Coefficient matrix: ", coefficientMatrix[i])
    print("Constant matrix: ", constantMatrix[i])
    print("Solution: ", gaussSeidel(coefficientMatrix[i], constantMatrix[i], 0.1, 15))
    print("")
```
