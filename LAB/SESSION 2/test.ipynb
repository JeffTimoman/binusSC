```python
x1 = x2 = x3 = 0

count = 1

while count != 100:
    y1 = (3 * x2 - 6 * x3 + 8.6)/10
    y2 = (9 * x3 - x1 + 0.8)/15
    y3 = (6 * x2 + x1 + 6.6)/8

    x1 = y1; x2 = y2; x3 = y3
    print(f"Iteration {count} : {x1} {x2} {x3}")
    count += 1

print(f"Final result : {x1} {x2} {x3}")
```
