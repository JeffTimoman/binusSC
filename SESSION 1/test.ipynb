```python

def sci_notation(number):
    number = float(number)
    if number == 0:
        return '0'
    else:
        return '{:.2e}'.format(number)
print(sci_notation(0.00000005))
```

```python
class Dragon:
    def __init__(self):
        self.name = 'Dragon'
        self.health = 100
        self.damage = 10
    
    def attack(self, target):
        target.health -= self.damage
        print(f"{self.name} attacked {target.name} for {self.damage} damage")
        print(f"{target.name} has {target.health} health left")
    
    def __str__(self):
        return f"{self.name} has {self.health} health and {self.damage} damage"

dragon1 = Dragon()
print(dragon1)
```

```python
class user:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"Name : {self.name}, Age : {self.age}"

class admin(user):
    def __init__(self, name, age, role):
        super().__init__(name, age)
        self.role = role
    
    def __str__(self):
        return f"Name : {self.name}, Age : {self.age}, Role : {self.role}"


admin1 = admin("Admin", 20, "Admin")
print(admin1)
```
